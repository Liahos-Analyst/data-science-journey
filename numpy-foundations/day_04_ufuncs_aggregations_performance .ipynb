```python
import numpy as np
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output
        
values = np.random.randint(1, 10, size=5)
%timeit compute_reciprocals(values)
 
```

### Intro to UFuncs

##### A universal function, or ufunc, is a function that performs element-wise operations on data in ndarrays. 
###### You can think of them as fast vectorized wrappers for simple functions that take one or more scalar values and produce one or more scalar results.

# Important UFunc Section

##### := is called the walrus operator. It assigns a value AND returns it in the same expression.

```python
arr1a = np.arange(1,11)
arr2a = np.arange (11,21)
arr3a = np.array([-3.1,  1, -2, 6.9, -7])
print(f'''
 Add: \n{np.add(arr2a, arr1a)}, \n{arr2a+arr1a}
\n Substract: \n{np.subtract(arr2a, arr1a)}, \n{arr2a-arr1a}
\n Multiply: \n{np.multiply(arr2a, arr1a)}, \n{arr2a*arr1a}
\n Divide: \n{np.divide(arr2a, arr1a)}, \n{arr2a/arr1a}
\n Floor: \n{np.floor_divide(arr2a, arr1a)}, \n{arr2a//arr1a}
\n Power: \n{np.power(arr2a, arr1a)}, \n{arr2a**arr1a}
\n Module: \n{np.mod(arr2a, arr1a)}, \n{arr2a%arr1a}
-------------------------------------------------------
np.negative flips the sign of every element. 
Positive becomes negative. Negative becomes positive.
-------------------------------------------------------
\n Negative: \n{np.negative(arr1a)}, { np.negative(arr1a) == -arr1a }

{np.negative(arr2a)}, { np.negative(arr2a) == -arr2a }
{np.negative(arr3a)}
\n Absolute: {np.abs(-arr2a)}, {np.abs(-arr1a)}, {np.abs(arr3a)}
\n Round: {np.rint(arr3a)}, {np.round(arr3a)}, { np.rint(arr3a) == np.round(arr3a) }
\n Round: { (a:= np.rint(arr3a) )}, { (b:= np.round(arr3a) )}, { a == b } #observe the behaviour of a:= and b:= here
\n Sign: {np.sign(arr3a)}
\n Floor: {np.floor(arr3a)}
\n Ceiling: {np.ceil(arr3a)}


\n Square: {np.square(arr1a)}, {np.square(arr2a)}
\n Square Root: {np.sqrt(arr1a)}, {np.sqrt(arr2a)}
\n Exponet: e raised to each element \n\n{np.exp(arr1a)}, \n\n{np.exp(arr3a)}
''')
```

```python
sales = np.array([
    [1000, 1500, 1200],   # Row 0 — Product A: North, Central, South
    [2000, 2500, 1800],   # Row 1 — Product B: North, Central, South
    [1500, 1000, 2200],   # Row 2 — Product C: North, Central, South
])

print(f'''
Sum: \t{sales.sum()}
Per Region:: {( a:= sales.sum(axis=0) )}, {a.shape} \'total per region\'
Per Product: {( b:= sales.sum(axis=1) )}, {b.shape} \'total per product\'

Region-wise 

max : { sales.max(axis=0) }
min : { sales.min(axis=0) }
avg: { sales.mean(axis=0) }

Product-wise

max : { sales.max(axis=1) }
min : { sales.min(axis=1) }
avg: { sales.mean(axis=1) }

Overall
max : { np.max(sales) }
min : { np.min(sales) }
avg: { np.mean(sales) }
''')
```

```python
%timeit sales.sum()
%timeit np.sum(sales)

'''
sales.sum():
  array.sum() → C function → result

np.sum(sales):
  np.sum() → type check → dispatch → array.sum() → C function → result
  ↑ extra steps here
-----------------------------------------------------------------------

Tip:

arr.sum()     → use this in performance-critical code
               tight loops, inference pipelines

np.sum(arr)   → use this when input type is uncertain
               might be a list, might be an array
               np.sum handles both gracefully
'''
```

#### ── PART 4 — Performance: proving it with numbers ────

```python
np.random.seed(20)
big_arr = np.random.randn(1_000_000)

# METHOD 1 — Python loop
def sum_loop(arr):
    total = 0.0
    for x in arr:
        total += x
    return total

# METHOD 2 — NumPy ufunc
def sum_numpy(arr):
    return np.sum(arr)

# Time them properly
start = time.perf_counter()
result_loop = sum_loop(big_arr)
time_loop = time.perf_counter() - start

start = time.perf_counter()
result_numpy = sum_numpy(big_arr)
time_numpy = time.perf_counter() - start

print(f"Loop result:  {result_loop:.4f}")
print(f"NumPy result: {result_numpy:.4f}")
print(f"Loop time:    {time_loop:.4f}s")
print(f"NumPy time:   {time_numpy:.6f}s")
print(f"Speedup:      {time_loop/time_numpy:.0f}x faster")
```

```python
import numpy as np

np.random.seed(42)

data = np.random.randint(1, 100, size=(5, 4, 3)).astype(np.float64)

data[:, :, 0] = data[:, :, 0] * 0.5      # delivery time: 0.5–50 hours
data[:, :, 1] = data[:, :, 1] * 1000     # cost: 1000–100000 PKR
data[:, :, 2] = (data[:, :, 2] / 10)     # rating: 0.1–10.0

cities  = ['Lahore', 'Karachi', 'Islamabad', 'Peshawar', 'Quetta']
weeks   = ['Week1', 'Week2', 'Week3', 'Week4']
metrics = ['Delivery_Time', 'Cost_PKR', 'Rating']

'''
data.shape = (5, 4, 3)
              ↑  ↑  ↑
           city wk met
'''
```

### Task 1 — ufuncs on real data

```python
ratings = data[:, :, 2]
print(f'''
 Ratings: \n{ratings}
\n round upto 1 decimal: \n{ np.round(ratings, 1) }
\n sqrt: \n{ np.sqrt(ratings) }
\n log: \n{ np.log(ratings) }
\n Cap: cap elements >= 8 to 8. If element is 9.2, will be replaced by 8. if element < 8, no changes
{ np.clip(ratings, 1, 8) } 
''')

# no dtype changes for any of above operation
```

### Task 2 — Aggregations with axis mastery

```python
total_cost = data[:, :, 1].sum()
avg_delievery_time = data[:, :, 0].mean(axis=1)
best_rating_week = data[:, :, 2].max(axis=0)
avg_metrics = data.mean(axis= (0, 1) )
print(f'''
Total Cost: {total_cost.astype(int)}
Average delivery time per CITY: {avg_delievery_time}, {avg_delievery_time.shape}
Best Rating per Week across all cities: {best_rating_week}, {best_rating_week.shape}
Average of each METRIC across everything: {avg_metrics}, {avg_metrics.shape}

''')
```

```python
def calculate_delivery_score(data):
    scores = []
    for city in range(data.shape[0]):
        for week in range(data.shape[1]):
            time   = data[city, week, 0]
            cost   = data[city, week, 1]
            rating = data[city, week, 2]
            score  = (rating * 10) - (time * 0.5) - (cost * 0.001)
            scores.append(score)
    return np.array(scores)

start = time.perf_counter()
result_slow = calculate_delivery_score(data)
time_slow = time.perf_counter() - start
#score is the most expensive term here cus for each element, it has to check dtype bfore doing nay operation.


def calculate_delivery_score_fast(data):
    time   = data[:, :, 0]
    cost   = data[:, :, 1]
    rating = data[:, :, 2]
    scores  = (rating * 10) - (time * 0.5) - (cost * 0.001)
    return scores.ravel()

start = time.perf_counter()
result_fast = calculate_delivery_score_fast(data)
time_fast = time.perf_counter() - start

print(np.allclose(result_slow, result_fast))  # True
print(f"Slow: {time_slow:.6f}s")
print(f"Fast: {time_fast:.6f}s")
print(f"Speedup: {time_slow/time_fast:.0f}x")
```

### Task 4 — Combining everything

```python
avg_rating_per_city = data[:, :, 2].mean(axis=1)
best_city_avg_index = np.argmax(avg_rating_per_city) # while np.max finds highest element in the array, np.argmax file the index of the highest element of the array

avg_cost_week = data[:, :, 1].mean(axis=0)
least_avg_cost_week = np.argmin(avg_cost_week)

r_min = ratings.min()
r_max = ratings.max()
normalized = (ratings - r_min) / (r_max - r_min)

bad_combos = (data[:, :, 0] > 20) & (data[:, :, 2] < 5.0)

print(f'''
Highest AVERAGE rating CITY: { cities[best_city_avg_index] }
Lowest AVERAGE cost WEEK: {weeks[least_avg_cost_week]}
Normalized ratings: \n{normalized}
{ normalized.min() }, { normalized.max() }
\n{bad_combos.sum()}, {bad_combos.dtype}, {bad_combos.sum().dtype}
''')
```
