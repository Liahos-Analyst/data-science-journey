#### Pseudorandom Number Generation

### Mental Note

###### rng.permutation (The Re-Arranger): 
It takes an existing array and returns a copy with the elements in a random order. It’s like shuffling a deck of cards.

##### rng.choice (The Sub-Sampler):  
It picks $N$ elements from a list. You can decide if you want to allow picking the same item twice (replace=True) or not (replace=False).

##### rng.integers (The Discrete Grid): 
It generates whole numbers within a specific range [low, high).

#### rng.standard_normal vs. rng.normal
These both deal with the Bell Curve (Gaussian Distribution), which is the \'Natural Law\' of data. Most errors, measurements, and physical traits follow this shape.

##### rng.standard_normal (The "Pure" Bell Curve)The Rule: 
The Mean ($\mu$) is always 0 and the Standard Deviation ($\sigma$) is always 1.Usage: It's the "raw" version. Engineers use it to generate "unit noise."

##### rng.normal (The "Custom" Bell Curve)The Rule: 
You define the Center (loc/mean) and the Spread (scale/std dev).Usage: Simulating real things. If you want to simulate "Human Heights," you don't start at 0. You start at 170cm with a spread of 10cm.

##### np.allclose(A @ x, b)
1. A @ x: It takes your starting vector $x$ and applies the transformation $A$ to see where it lands.
2. Comparison: It compares that landing spot to your target $b$, element by element.
3. Tolerance: It checks if the difference is smaller than a very tiny threshold (by default, $1 \times 10^{-8}$).
4. The Result: It returns True if every single element is "close enough," and False if even one element is significantly off.

#### ── PART 4 — Performance profiling ──────

##### %timeit — best for single operations in Jupyter 
Runs multiple times and reports mean ± std dev

##### time.perf_counter — best for larger blocks of code

```python
rng = np.random.default_rng(42)
X = rng.standard_normal((10_000, 100)) # equivalent to 'np.random.randn(10000, 100)'
w = rng.standard_normal(100)

def predict_loop(X, w):
    results = []
    for row in X:
        results.append( np.dot(row, w) )
    return np.array(results)

def predict_matmul(X,w):
    return X @ w

def predict_einsum(X, w):
    return np.einsum('ij,j->i', X, w)

# Time all three
start = time.perf_counter()
r1 = predict_loop(X, w)
t1 = time.perf_counter() - start

start = time.perf_counter()
r2 = predict_matmul(X, w)
t2 = time.perf_counter() - start

start = time.perf_counter()
r3 = predict_einsum(X, w)
t3 = time.perf_counter() - start

print(f"Loop:    {t1:.4f}s")
print(f"Matmul:  {t2:.6f}s")
print(f"Einsum:  {t3:.6f}s")
print(f"Matmul speedup: {t1/t2:.0f}x")

# Verify all give same result
print(np.allclose(r1, r2)) 
print(np.allclose(r1, r3))
```
