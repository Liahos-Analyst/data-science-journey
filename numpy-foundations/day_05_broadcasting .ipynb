### Rules of Broadcasting
Broadcasting in NumPy follows a strict set of rules to determine the interaction between the two arrays:
###### • Rule 1: If the two arrays differ in their number of dimensions, the shape of the one with fewer dimensions is padded with ones on its leading (left) side.
###### • Rule 2: If the shape of the two arrays does not match in any dimension, the array with shape equal to 1 in that dimension is stretched to match the other shape.
###### • Rule 3: If in any dimension the sizes disagree and neither is equal to 1, an error is raised.

#### Broadcasting example 1

```python
m = np.ones((2,3))
a = np.arange(3)
print(f'''
m: {m.shape}, {m.dtype} \n{m}\n 
a: {a.shape}, {a.dtype} \n{a}
''')

'''
By Rule 1:
M.shape -> (2, 3)
a.shape -> (1, 3)

By Rule 2:
M.shape -> (2, 3)
a.shape -> (2, 3)

'''
print(f' np.add(m+a): {( c:= m+a ).shape}, {c.dtype} \n{c} ')
```

#### Broadcasting example 2

```python
a = np.arange(3)[:, np.newaxis]
b = np.arange(3)
print(f'a: {a.shape} \n{a} \n\nb: {b.shape} \n{b}')

'''
Rule 1:
a.shape -> (3, 1)
b.shape -> (1, 3)

Rule 2:
a.shape -> (3, 3)
b.shape -> (3, 3)

'''

print(f'\na+b: {(a+b).shape} \n{a+b}')
```

#### Broadcasting example 3

```python
m = np.ones((3,2))
n = np.arange(3)

print(f'm: {m.shape} \n{m} \n\nn: {n.shape} \n{n}')
'''
Rule 1:
M.shape -> (3, 2)
a.shape -> (1, 3)

Rule 2:
M.shape -> (3, 2)
a.shape -> (3, 3)

Rule 3:
the final shapes do not match, so these two arrays are incompatible
'''
m+n
```

#### ── PART 2 — Shapes that work and shapes that fail ─────

```python
a = np.ones((3, 4))
b = np.ones(4)

'''
a: (3,4)
b: (4,)

Rule 1:
a: (3,4)
b: (1,4)

Rule 2:
a: (3,4)
b: (3,4)

'''
print((a+b).shape)
```

```python
e = np.ones((3, 4))
f = np.ones((3,)) 

'''
e: (3,4)
f: (3, )

Rule 1:
e: (3,4)
f: (1,3)

Rule 2:
e: (3,4)
f: (3,3)

'''
print((e + f).shape)
```

#### ── PART 3 — The reshape trick for column operations ─────

```python
city_discount = np.array([0.10, 0.20, 0.15])
city_discount_col = city_discount[:, np.newaxis]
print( city_discount.shape )

result_col = sales * (1 - city_discount)

'''
sales:         (3,3)
city_discount: (3,1)

Rule 2:
sales:         (3,3)
city_discount: (3,3)

'''
print( result_col, np.shares_memory(city_discount, city_discount_col) )
```

```python
metric_means = np.mean(data, axis = (0,1))

print( metric_means.shape, metric_means)
data_sub = data - metric_means
print('\n\n', data_sub.shape, np.shares_memory(data, data_sub), '\n', data_sub)
# shares_memory(data, data_sub) False — arithmetic always creates new array

'''
data:         (5,4,3)
metric_means: (3,)

Rule 1: 
data:         (5,4,3)
metric_means: (1,1,3)

Rule 2:
data:         (5,4,3)
metric_means: (1,4,3)
metric_means: (5,4,3)

'''

# 3. Each city has a performance multiplier:
city_multipliers = np.array(
    [1.1, 0.9, 1.0, 1.2, 0.8])  # shape (5,)


'''
data[:, :, 0]:    (5,4)
city_multipliers: (5,)

Rule 1:
data[:, :, 0]:    (5,4)
city_multipliers: (1,5)

Rule 3 applies here, can't multiple

But with reshape, it can happen

data[:, :, 0]:        (5,4)
city_multipliers_col: (5,1)

Rule 2:
data[:, :, 0]:        (5,4)
city_multipliers_col: (5,4)

'''


city_multipliers_col = city_multipliers[:, np.newaxis]
city_delivery_time = data[:, :, 0] * city_multipliers_col
print('\n', np.shares_memory(city_multipliers_col, city_multipliers), '\n', city_delivery_time )

print('\n\n', city_delivery_time )
```
