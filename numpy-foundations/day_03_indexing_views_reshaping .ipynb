## Boolean Indexing

```python
names = np.array( ['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'] )
data = np.array([ [4, 7], [0,2], [-5,6], [0,0], [1,2], [-12,-4], [3,4] ])
names
```

```python
data
```

```python
bol = names == 'Bob'
print(bol)
data[bol]
```

```python
bol0 = names == 'Joe'
print(bol0)
data[bol0]
```

```python
data[bol, 1:]
```

```python
data[bol, 1]
```

```python
bol = names != 'Bob'
print(bol)
data[bol]
```

```python
bol = ~(names == 'Bob')
print(bol)
data[bol]
```

```python
data[~bol]
```

```python
data[bol, 1]
```

```python
data[bol, 1:]
```

```python
mask = (names == 'Bob') | (names == 'Will') # | is used as OR operator here
data[mask]
```

```python
mask = (names == 'Bob') & (names == 'Will') # & is used as AND operator here
data[mask]
```

```python
print(data)
data[data > 2]
```

```python
data[names != 'Joe'] = 7
data
```

```python
ravelled  = matrix2.ravel()
flattened = matrix2.flatten()
#Both take a multi-dimensional array and return a flat 1D array.

print(ravelled, np.shares_memory(matrix2, ravelled))    # True
print(flattened, np.shares_memory(matrix2, flattened))   # False
```
