## Guided Task

 ── PART 1 — Basic arithmetic between arrays ──────────

```python
a = np.array([10, 20, 30, 40], np.int16)
b = np.array([1, 2, 3, 4], np.int16)
add = a+b
sub = a-b
mul = a*b
div = a/b
floor = a//b
square = a**2
power = a**b
print('Add:         ',add, add.dtype)
print('Substract:   ', sub, sub.dtype)
print('Multiply:    ', mul, mul.dtype)
print('Divide:      ', div, div.dtype)
print('Floor:       ', floor, floor.dtype)
print('Square:      ', square, square.dtype)
print('Power:       ', power, power.dtype)
```

 ── PART 2 — Scalar operations ──────────

```python
# audit the data type
sales = np.array([1000, 1500, 1200, 1800], np.int32)
with_tax = sales * 1.15 # Apply 15% tax,
print(with_tax, with_tax.dtype)

with_tax = (sales * 1.15).astype(np.float32) # Apply 15% tax,
print(with_tax, with_tax.dtype)

with_tax_f4 = with_tax.astype('f4')
print(with_tax_f4, with_tax_f4.dtype)

with_tax_f32 = sales * np.float32(1.15) # dtype should have been float32 but technical error
print(with_tax_f32, with_tax_f32.dtype)

with_tax_f32 = sales.astype(np.float32) * np.float32(1.15) # np.float32 = 'f4'
print(with_tax_f32, with_tax_f32.dtype)
```

 ── PART 3 — The silent upcasting rule ─────────

```python
x = np.arange(6,)
y=np.arange(7,13).astype(np.float64)
z = x+y
print('x:', x.dtype)
print('y:', y.dtype)
print('z:', z, z.dtype)

# int8 + float64 = float64, numpy always upcast to highest precision
```

# Project

```python
import numpy as np
revenue = np.array([320, 415, 390, 502, 478, 610])
costs = np.array([210, 280, 260, 310, 305, 390])
```

### Task 1 — Profit analysis
Calculate and print:

Monthly profit for each month (one operation)
Total profit across all 6 months
Average monthly profit
Best month profit and worst month profit

No loops. One NumPy operation per answer.

```python
profit = revenue - costs
print(f'Profit for each month: {profit}')

bi_annualy_profit = profit.sum()
print(f'Total profit across all months: {bi_annualy_profit}')

# avg_monthly_profit = bi_annualy_profit/revenue.size
avg_monthly_profit = np.round(profit.mean()).astype('i4')
print(f'Average Monthly profit: {avg_monthly_profit}')

print(f'Highest monthly profit: {profit.max()}')
print(f'Lowest monthly profit: {profit.min()}')
```

### Task 2 — The silent upcast trap
fix it so the output is float32 instead of float64. Show the memory saving using nbytes

```python
tax_rate = 0.17
revenue_after_tax = revenue * tax_rate
bytes_used = revenue_after_tax.nbytes
print(revenue_after_tax.dtype)
print(bytes_used)
# be default tax_rate is float64 or 'f8' and its gonna take almost double the memory space allocated

revenue_after_tax_f4 = np.float32(revenue) * np.float32(tax_rate)
bytes_new = revenue_after_tax_f4.nbytes
print(revenue_after_tax_f4.dtype)
print(bytes_new)

print(f'Saved Memory: {bytes_used-bytes_new}')
```

### Task 3 — Boolean filtering
Using the revenue and costs arrays:

Find which months had profit greater than 150 (thousand PKR)
Count how many such months there were
Find what percentage of months were high-profit months

All in NumPy. No loops. No if statements.

```python
threshold = profit > 150
high_profit_months = profit[threshold]
count = threshold.sum()
print(f'Profit Months > 150k: {high_profit_months}')
print(f'Total Months > 150k: {count}')

percentage = ( count/len(profit) )*100
print(f"Percentage: {percentage}%") 
```
