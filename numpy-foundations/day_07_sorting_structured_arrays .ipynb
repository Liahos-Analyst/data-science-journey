### Unique and Other Set Logic

##### np.union1d performs a Set Union.

###### 1. It gathers all elements from both arrays.

###### 2. It De-duplicates them (removes repeats).

###### 3. It Sorts them.

##### np.in1d(x, y) 

###### Compute a Boolean array indicating whether each element of x is contained in y

```python
''''
Company made 2 campigns to increase their sales and see the value it brings. 
Campaign_members has data of participants and 
Employ_id has data of employee ids involved in the campaign.
'''

campaign1_members = np.array(['Bob', 'Joe', 'Bob', 'Will', 'Joe', 'Bob'])
campaign2_members = np.array(['Joe', 'King', 'Bob', 'King'])
blacklisted_members = np.array(['King'])

employ_id_1 = np.array([1, 2, 3, 2, 3, 4, 5, 5, 4, 6])
employ_id_2 = np.array([1, 3, 5, 6, 2, 3, 5, 8, 9, 10])

all_members = np.union1d(campaign1_members, campaign2_members)
both_campaign_members = np.intersect1d(campaign1_members, campaign2_members)
mask = np.in1d(all_members, blacklisted_members) # to remove blacklisted members

all_employees = np.union1d(employ_id_1, employ_id_2)
both_campaign_employees = np.intersect1d(employ_id_1, employ_id_2)

print(f'''
Participants of both campaigns: {all_members}
Loyal Customers:{both_campaign_members}
Unique Names: {np.unique(campaign1_members)}
Verified Participants: {all_members[~mask] }

All involved Employees: {all_employees}
Special Reward Employees:{both_campaign_employees}
Unique Employees: {np.unique(employ_id_1)}

''')
```

#### ── PART 1 — Sorting ─────

```python
arr = np.array([3, 1, 4, 1, 5, 9, 2, 6])
sorted_arr = np.sort(arr)
top_arr = np.argsort(arr)[::-1]
print(f'''
array: {arr}
sorted array: {sorted_arr}, {np.shares_memory(arr, sorted_arr)}
top array: {arr[top_arr]}, {np.shares_memory(arr, top_arr)}
''')

arr_copy = arr.copy()
arr_copy.sort()
print(arr_copy, np.shares_memory(arr, arr_copy) )

scores = np.array([85, 92, 78, 95, 88])
rank_indicies = np.argsort(scores)[::-1]
print('\nRanked scores:', rank_indicies)

names = np.array(['Ali', 'Sara', 'Umar', 'Zara', 'Bilal'])
print(f'{names[rank_indicies]} \n{scores[rank_indicies]}')


matrix = np.array([[3, 1, 4],
                   [1, 5, 9],
                   [2, 6, 5]])
print(f'''
sort each COLUMN \n{np.sort(matrix, axis=0)}

sort each ROW \n{np.sort(matrix, axis=1)}
''')
```

#### ── PART 2 — Searching ────

```python
data = np.array([1, -2, 3, -4, 5, -6])

#where(condition, if_true, if_false)
result = np.where(data >0, data, 0)
print(result)

indicies = np.where(data>0)
final = result[indicies] # gonna store/print non-zeros elemets in result array
print( final, np.shares_memory(final, result) )

# np.searchsorted — find insertion point in sorted array
sorted_data = np.array([10, 20, 30, 40, 50])
print(f'''
{ np.searchsorted(sorted_data, 25) }
{ np.searchsorted(sorted_data, [25, 35]) }
''')

# np.argmax / np.argmin — index of max/min
print( np.argmax(sorted_data), np.argmin(sorted_data) )
```

##### np.searchsorted(data, 25) $\rightarrow$ 2
Value 25 is greater than 20 (Index 1) but less than 30 (Index 2).To keep the order, 25 "pushes" 30 aside and takes Index 2.

##### np.searchsorted(data, [25, 35]) $\rightarrow$ [2, 3]
For 25: It belongs at Index 2.For 35: It belongs between 30 (Index 2) and 40 (Index 3). It takes Index 3.

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([3, 4, 5, 6, 7])

print(f'''
unique a: {np.unique(a)}
unique b: {np.unique(b)}

union (a, b): {np.union1d(a, b)}
intersect(a, b): {np.intersect1d(a,b)}

'Compute a Boolean array indicating whether each element of a is contained in b'
bool: {np.in1d(a, b)} 

''')
```

```python
arr = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5, 3])
'''
NumPy scans your array and builds a Reference Key (unique) and a Scorecard (counts).

Task 1 (Unique): It identifies every distinct number, removes duplicates, and sorts them.

Task 2 (Counts): It goes back through the original array and tallies how many times each number in your "Reference Key" appeared.
'''
unique, counts = np.unique(arr, return_counts=True)
print(unique)    # [1 2 3 4 5 6 9]
print(counts)    # [2 1 2 1 2 1 1] — how many of each

duplicate = counts>1
print(duplicate)
print(unique[duplicate])
```

#### ── PART 4 — Missing data with np.nan ───

```python
data = np.array([1.0, np.nan, 3.0, np.nan, 5.0], 'f8')
print(f'''
\t***Regular Operations***

Mean: {np.mean(data)}
Sum: {np.sum(data)}
bool: {np.isnan(data)}
nan: {np.isnan(data).sum()}

\t***nan-safe operations***
Mean: {np.nanmean(data):.0f}
Sum: {np.nansum(data):.1f}
Min: {np.nanmin(data)}
Max: {np.nanmax(data)}

''')

mask = np.isnan(data)
print(data[mask])
data[mask] = np.nanmean(data)
print(data)
```
