````python
'''
np.ones() / np.ones_like() — The Ones Filler

Creates an array filled entirely with 1s. 
ones_like copies the shape and dtype of an existing array and fills it with 1s.
'''
````

````python
np.ones(6)
````

````python
z = np.ones((2, 3,3))*10
z
````

````python
np.ones_like(z)
````

````python
'''
np.zeros() / np.zeros_like() — The Zeros Filler

Identical to ones but fills with 0s. 
Most common array initializer in ML — model weights often start at zero.
'''
````

````python
np.zeros(6)
````

````python
np.zeros((3, 2, 2))
````

````python
z
````

````python
np.zeros_like(z, np.int8)
````

````python
'''
np.empty() / np.empty_like() — The Uninitialized Buffer

Allocates memory but does not fill it with anything. 
Whatever random bytes were sitting in that RAM location — that is what you get. 
Fastest creation function because it skips the fill step.
'''
````

````python
np.empty((2, 2, 4)) # fills with random values
````

````python
np.empty_like(z) # fills with random values 
````

````python
# ── PART 1 — Your first array ─────────────────────────────────

python_prices = [500, 1200, 850, 750]
numpy_prices = np.array(python_prices)
print(numpy_prices)
print(type(numpy_prices))
````

````python
# ── PART 2 — The metadata contract ───────────────────────────
print(len(numpy_prices))
print(numpy_prices.dtype)
print(numpy_prices.shape)
print(numpy_prices.ndim)
print(numpy_prices.size)
print(numpy_prices.itemsize)
print(numpy_prices.nbytes)
````

````python
# ── PART 3 — Controlling the dtype ───────────────────────────

# float64 — decimal numbers, 8 bytes each
prices_f64 = np.array(python_prices, np.float64)
print(prices_f64)
print(prices_f64.itemsize)
print(prices_f64.nbytes)

# float32 — less precision, 4 bytes each — used in ML models
prices_f32 = np.array(python_prices, np.float32)
print(prices_f32)
print(prices_f32.itemsize)
print(prices_f32.nbytes)

'''Same values. Half the memory and low precision
1 million rows × 100 features = 800MB in float64
                              = 400MB in float32
This is why dtype choice matters in production.
'''
````

````python
# ── PART 4 — Creating arrays without typing every number ─────
zeros =  np.zeros(6, np.int8)
ones = np.ones(6, np.int8)
weekdays = np.arange(1, 8)
percentages = np.linspace(0,100, 5)

print(zeros)
print(ones)
print(weekdays)
print(percentages)
````

````python
# ── PART 5 — The operation that makes NumPy worth using ───────
prices = [1000, 900, 1100, 3000, 700]
original_prices = np.array(prices)
sale = .25
sales = original_prices*sale
discounted_prices = original_prices - sales

print(original_prices)
print(discounted_prices)
````

````python
**The scenario:** You are a data analyst at a company in Lahore. Here is one week of daily website visitors:
```
Monday: 1240  Tuesday: 980   Wednesday: 1560
Thursday: 1100  Friday: 1890  Saturday: 2340  Sunday: 760
````

````python
visitors = [1240, 980, 1560, 1100, 1890, 2340, 760]
````

Create a NumPy array from this data. Print all six metadata values: dtype, shape, ndim, size, itemsize, nbytes.

````python
weekly_visitors = np.array(visitors)
print(f'Weekly Visitors: {weekly_visitors}')
print(f'Data Type: {weekly_visitors.dtype} ')
print(f'Shape: {weekly_visitors.shape}')
print(f'Dimensions: {weekly_visitors.ndim}')
print(f'Size: {weekly_visitors.size}')
print(f'Size per Item: {weekly_visitors.itemsize}' )
print(f'Total Bytes: {weekly_visitors.nbytes}')
````

Task 2 — Summary statistics, one line each

Using your visitors array, find and print:

Total visitors for the week

Average daily visitors

Highest single day

Lowest single day

Difference between highest and lowest

````python
print(f'Total visitors for the week: {weekly_visitors.sum()}')
print(f'Average Daily Visitors: {weekly_visitors.mean()}')
print(f'Highest single day visitor: {weekly_visitors.max()}')
print(f'Lowest single day visitor: {weekly_visitors.min()}')
print(f'Difference between highest and lowest: {weekly_visitors.max() - weekly_visitors.min()}')
````

Task 3 — dtype experiment

Create two new arrays from the same visitor data:

One with dtype=np.float32

One with dtype=np.int16

````python
weekly_visitors_f32 = weekly_visitors.astype(np.float32)
weekly_visitors_i16 = weekly_visitors.astype(np.int16)

print(weekly_visitors_f32)
print(weekly_visitors_i16)
````

Your manager wants projected visitor numbers if the site grows 10% per month for 6 months. 
Start from your weekly average in Task 2.

````python
weekly_visitor_avg = weekly_visitors.mean() 
months = np.arange(1, 7)
projection = np.round( weekly_visitor_avg * (1.10 ** months) ).astype(int)
print(projection)
````
