```python
import numpy as np
import pandas as pd

balances = pd.Series( 
    {'Ali': 15000, 'Sara':42000, 'Usman':8500, 'Hina':61000}, 
    dtype=np.int32
)

balances.index.name = 'Name'
balances

print(f'''
Shape: {balances.shape}
Index: {balances.index.tolist()}

Ali's balance   : {balances['Ali']:,}  go to the row labeled 'Ali'." (Label-based)
First entry     : {balances.iloc[0]:,}  go to the very first row, regardless of its name." (Position-based)

After 10% raise: \n{balances*1.10}
''')

new_balance = balances.iloc[:]
np.shares_memory(balances, new_balance)
```

# Project

##### A sabzi mandi in Lahore tracks daily stock (kg) for 5 vegetables: Tomato 120, Onion 340, Potato 210, Garlic 95, Ginger 60.


1. Build the Series with explicit dtype=np.int32

2. Print only vegetables with stock below 150 kg (boolean mask)

3. Add a 20% emergency restock to low item — store as restocked

4. Print the name and value of the highest-stocked item after restocking

```python
inventory = pd.Series(
    {'Tomato': 120, 'Onion': 340, 'Potato': 210, 'Garlic': 95, 'Ginger': 60}
)

inventory.dtype
inventory.astype('i4')
vegetables = inventory.copy()

low_items = vegetables<150
require_restock = inventory[low_items]

np.shares_memory(vegetables, inventory)
np.shares_memory(vegetables, require_restock)

restocked = require_restock*1.2

vegetables[low_items] = restocked
vegetables

'np.argmax(): Returns the Integer Position'
'obj.idxmax(): Returns the Index Label'

top_quantity = vegetables.idxmax() 
print('Overstocked Item:', top_quantity)
```

```python
import time

big_data = pd.Series(np.random.randint(50, 500, size=1_000_000))

# --- Measure Approach: Loop ---
start_1 = time.perf_counter()
worst = [x for x in big_data if x < 150]
time_1 = time.perf_counter() - start_1

# --- Measure Approach:Vectorized ---
start_2 = time.perf_counter()
best = big_data[big_data < 150]
time_2 = time.perf_counter() - start_2

# --- THE COMPARISON ---
print(f"Loop Time: {time_1:.6f} seconds")
print(f"Vector Time: {time_2:.6f} seconds")
print(f"Speedup: {time_1 / time_2:.2f}x faster!")
```
