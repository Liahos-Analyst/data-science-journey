```python
learnt = {
    'lower()': 'its gonna lower cap the desired word', 
    'upper()': 'its gonna upper cap the desired word',
    'title()': 'its gonna first letter capital of the desired word',
    'lstrip()': 'its gonna remove whitespaces from left',
    'rstrip': 'its gonna remove whitespaces from right',
    'strip': 'its gonna remove whitespaces from either side of the desired word',
    'l_strip()': 'its gonna remove whitespaces from left',
    'tutle()': 'its gonna first letter capital of the desired word',
    'set()': 'remove  duplicates from dictionary',
}

v = 0
for value in sorted(learnt.values()):
    v +=1
    print(f'{v}. {value}')

v=0
print('\n\n\n')
for value in sorted(set(learnt.values())):
    v+=1
    print(f'{v}. {value}')
```
