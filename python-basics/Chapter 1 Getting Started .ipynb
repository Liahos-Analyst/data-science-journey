```python
fav_language =' python '
print( 'Main String: ')
print("'" + fav_language + "'")
print('original length of the string: ',  len(fav_language))

print( '\n\nAfter using rstrip'+ ':')
print("'"+ fav_language.rstrip()+ "'") #rstrip removes whitescpace from Right only
print('length of the string with rstrip: ', len((fav_language.rstrip())))

print( '\n\nAfter using lstrip'+ ':')
print("'"+ fav_language.lstrip()+ "'") #rstrip removes whitescpace from left only
print('length of the string with lstrip: ', len((fav_language.lstrip())))

print( '\n\nAfter using strip'+ ':')
print("'" + fav_language.strip() + "'") #rstrip removes whitescpace from both sides
print('length of the string with strip: ', len((fav_language.strip())))
```
