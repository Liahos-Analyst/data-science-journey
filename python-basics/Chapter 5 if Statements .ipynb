```python
import random
cars = ['audi', 'bmw', 'subaru', 'toyota', 'hyundai']

print('''\t\t***Hey, welcome to the betiing game.***\n
                        ------- Good Luck -------\n''')
print('''Instruction: 
1. You can make only 5 consective Flase prediction today.
2. Press Q to quit the game.\n''')

wrong = 0; true =0; 
car = random.choice(cars)
while True:
    predict = input('Prediction: Which car is gonna win today?').lower().strip() 
    if predict == car:
        wrong = 0
        true +=1
        print(f'       ***Result: {car.lower()==predict} prediction.*** \n\t Limit has been reset to 0.\n \n')
        again = input('Would you like to play again? (YES/NO): ').lower().strip()
        if again == 'no':
            print('\n\n   ---You have selected NO. The game has ended.---  ')
            break
        if again == 'yes':
            car = random.choice(cars)
        if again == 'q':
            print(f'\nYou have pressed {again.upper()}. Game is quitting...')
            break          
    elif predict == 'q':
        print(f'\nYou have pressed {predict.upper()}. Game is quitting...')
        break       
    elif predict != car:
        wrong +=1
        true = 0
        if wrong <= 3:
            print('       ***Result: Wrong Prediction.***')
        elif true == 0 and wrong > 2:
            print('       ***Result: Wrong Prediction.***')
            sure = input('Are you sure you wanna continue? (YES/NO)')
            if sure.lower().strip() == 'no' and wrong !=5:
                print('\n\n   ---You have selected NO. The game has ended.---  ')
                break
        elif wrong > 4:
            print('\nLast Try.')
            
        if wrong == 5:
            print('\nWarning: You have reached your limit for today. \n\t ------ENDED-----')
            break
        print(f'\t    Remaining {5-wrong} tries.')
        
```

5-8. Hello Admin: Make a list of five or more usernames, including the name
'admin'. Imagine you are writing code that will print a greeting to each user
after they log in to a website. Loop through the list, and print a greeting to
each user:
• If the username is 'admin', print a special greeting, such as Hello admin,
would you like to see a status report?
• Otherwise, print a generic greeting, such as Hello Eric, thank you for logging
in again.

```python
import random
usernames = ['admin', 'eric', 'jones', 'snow', 'white', 'queen']
for value in range(0, len(usernames)):
    username = random.choice(usernames)
    #username = input()
    print(f'\n{username.title()}')
    if username.lower() == usernames[0]:
        print(f'Hello {username.title()}, would you like to see a status report?')
    elif username.lower() in usernames[1:]:
        print(f'Hello {username.title()}, thank you for logging in again.')
    elif username not in usernames:
        print('Access is denied')
```
